```python
third_lowest = df.sort_values(['budget']).iloc[2]['original_title']
third_highest = df.sort_values(['budget']).iloc[-3]['original_title']
third_lowest, third_highest 
```

```python
movies_between_2000_and_2005 = df[df['release_year'].isin([2000, 2001, 2002, 2003, 2004, 2005])]
```

```python
no_of_words = 0
for i in range(movies_between_2000_and_2005.shape[0]):
    words_list = movies_between_2000_and_2005['original_title'].values[i].split(' ')
    no_of_words = no_of_words + len(words_list)
    
ans2 = no_of_words/movies_between_2000_and_2005.shape[0]
ans2
```

```python
vin_diesel_df = df[df['cast'].str.contains('Vin Diesel')]
emma_watson_df = df[df['cast'].str.contains('Emma Watson')]
#for i in range(df.shape[0]):
    
```

```python
genres_and_count = {}
for i in range(vin_diesel_df.shape[0]):
    genres = vin_diesel_df['genres'].values[i].split('|')
    for genre in genres:
        try:
            count = genres_and_count[genre]
            genres_and_count[genre] = count + 1
        except:
            genres_and_count[genre] = 1
            
genres_and_count
```

```python
genres_and_count = {}
for i in range(emma_watson_df.shape[0]):
    genres = emma_watson_df['genres'].values[i].split('|')
    for genre in genres:
        try:
            count = genres_and_count[genre]
            genres_and_count[genre] = count + 1
        except:
            genres_and_count[genre] = 1
            
genres_and_count
```
